**Perform web-scraping of the proposed site and retrieve the list of city/town status, population, and county using the BeautifulSoup library**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_cities_and_towns_in_Kenya_by_population'
```

```python
response = requests.get(url)
```

```python
status_check = response.status_code
```

```python

'''
This code intends to establish whether a connection was triggered and data received and stored in our variable response
'''
if status_check == 200:
  print('Access approved and data received')
elif status_check == 404:
  print('Access denied')
else:
  print('Error in code')
```

```python
soup = BeautifulSoup(response.text, 'html')
```

```python
table = soup.find_all('table')[2]
```

```python
table_titles = table.find_all('th')
```

```python
table_titles_headers = [title.text.strip() for title in table_titles]
```

```python
table_titles_headers
```

```python
df = pd.DataFrame(columns = table_titles_headers)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  length = len(df)
  df.loc[length] = individual_row_data
```
